# Exploratory Data Analysis

```python
import numpy as np
import sys
sys.path.insert(0, '..')
import paths
from os.path import join
import time
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
start = time.time()
xyzrgb = np.loadtxt(join(paths.DATA_DIR, 'office_1.txt'))
end = time.time()
print('Loading {} xyzrgb points takes {:.02f} seconds.'.format(xyzrgb.shape[0], end-start))
```

```python
xyzrgb.shape
```

```python
for i in range(xyzrgb.shape[1]):
    print('Min for column {} is {}'.format('xyzrgb'[i], xyzrgb[:,i].min()))
    print('Max for column {} is {}'.format('xyzrgb'[i], xyzrgb[:,i].max()))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
start = time.time()

skip_factor = 100
ax.scatter(xyzrgb[::skip_factor, 0], xyzrgb[::skip_factor, 1], xyzrgb[::skip_factor, 2], s = 1)
plt.show()
print('Plotting {} points takes {} seconds.'.format(int(xyzrgb.shape[0]/skip_factor), time.time()-start))
```
